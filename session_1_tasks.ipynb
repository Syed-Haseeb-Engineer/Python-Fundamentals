### Q3:- Take 2 numbers as input from the user.Write a program to swap the numbers without using any special python syntax.

````python
# Write your code here
print("--- Program to Swap two numbers using temp variable ---")
first = int(input("Enter the first integer number : "))
second = int(input("Enter the second integer number : "))
print("Numbers Before Swapping are First:", first ,"Second:", second)
temp = first
first = second
second = temp
print("Numbers After Swapping are First:", first ,"Second:", second)
````

### Q4:- Write a program to find the euclidean distance between two coordinates.Take both the coordinates from the user as input.

````python
# Write your code here
x1 = int(input("Enter x1:"))
y1 = int(input("Enter x2:"))
x2 = int(input("Enter y1: "))
y2 = int(input("Enter y2:"))

#Euclidean Distance is use for calculating the length of the line
#Formula For Euclidean Distance
euclidean_dis = ((x2-x1)**2 + (y2-y1)**2)**0.5
print(euclidean_dis)
````

### Q5:- Write a program to find the simple interest when the value of principle,rate of interest and time period is provided by the user.


````python
# Write your code here
print("---Program to calculate simple interest ---")
principle = int(input("Enter the principle Amount: "))
interest_rate = int(input("Enter the Rate of Inrest:"))
period = int(input("Enter the Time/period in yearly basis: "))
si = (principle*interest_rate*period)/100
print("For", principle, "Amount", interest_rate, "Interest", period, "Period The simple Interest is", si)
````

### Q6:- Write a program that will tell the number of dogs and chicken are there when the user will provide the value of total heads and legs.

For example:
Input:
heads -> 4
legs -> 12
<br>
Output:
dogs -> 2
chicken -> 2





````python
# Write your code here
total_heads = int(input("Enter the total number of heads"))
total_legs = int(input("Enter the total number of legs"))
dogs = (total_legs - (2 * total_heads)) // 2
chickens = total_heads - dogs
print("Dogs :", dogs, "Chicken", chickens)
````

### Q8:- Given the first 2 terms of an Arithmetic Series.Find the Nth term of the series. Assume all inputs are provided by the user.

````python
# Write your code here
first = int(input("Enter first number"))
second = int(input("Enter second number"))
nth_term = int(input("Enter nth term:"))
d = second - first
an = (first + (nth_term-1) * d)
print(an)
````

### Q9:- Given 2 fractions, find the sum of those 2 fractions.Take the numerator and denominator values of the fractions from the user.

### Q10:- Given the height, width and breadth of a milk tank, you have to find out how many glasses of milk can be obtained? Assume all the inputs are provided by the user.



Input:<br>
Dimensions of the milk tank<br>
H = 20cm, L = 20cm, B = 20cm
<br><br>
Dimensions of the glass<br>
h = 3cm, r = 1cm

````python
# Write your code here
H = int(input("Enter the hight of tank:"))
L = int(input("Enter the length of Tank:"))
B = int(input("Enter the Breadth of Tank:"))

h = int(input("Enter the hight of glass: "))
r = int(input("Enter the radius of glass: "))
tank_volume = H*L*B
glass_volume = (3.14 * (r**2)  * h)
total_glass = (tank_volume//glass_volume)
print("The {} of glass".format(total_glass))
````
